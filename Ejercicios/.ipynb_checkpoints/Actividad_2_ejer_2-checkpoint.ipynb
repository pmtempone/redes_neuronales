```python
radar -= radar.min()  # equivalent to df = df - df.min()

radar /= radar.max()  # equivalent to df = df / df.max()

print(radar.head())
```

# b.elija 3 valores de alfa

```python
alfa1 = 0.1
alfa2 = 0.5
alfa3 = 0.8
```

# c. Pruebe de entrenar el clasificador usando logsig y tansig como función de activación

```python
x = np.array(radar)
```

```python
T = x[:,2]
P = x[:,:2].T
```

```python
#Alfa 1 logsig

salidas = []
salidas_parcial = (train(P,T,alfa1,1000,0.1,"logsig",True))

salidas.append(salidas_parcial)
```

```python
#Alfa 2 logsig

salida_parcial = ([train(P,T,alfa2,50,0.1,"logsig",True)])

salidas.append(salida_parcial)
```

```python
#Alfa 3 logsig

salida_parcial = (train(P,T,alfa3,1000,0.1,"logsig",True))

salidas.append(salidas_parcial)
```

```python
#Alfa 1 tansig

salidas_parcial = (train(P,T,alfa1,1000,0.1,"tansig",True))

salidas.append(salidas_parcial)
```

```python
#Alfa 2 tansig

salidas_parcial = (train(P,T,alfa2,1000,0.1,"tansig",True))

salidas.append(salidas_parcial)
```

```python
#Alfa 3 tansig

salidas_parcial = (train(P,T,alfa3,50,0.1,"tansig",True))

salidas.append(salidas_parcial)
```

El valor de los 6 entrenamientos muestra que en la función tansig con un alfa elevado le cuesta converger a la solución.
Y con las otras soluciones converge en pasos similares poniendo la cota en 0.1.
